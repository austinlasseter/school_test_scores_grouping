Python question. 3 columns w student level data: school code, test 1 and test 2.  I want to get the highest score between test 1 and test 2 and the take the max by school level.  

```python
import pandas as pd
import random
```

```python
# create a function for mock dataframes
def make_school(school_code):
    school_code=[school_code]*5
    student_codes =[random.randint(100,500) for i in range(0,5)]
    test1 = [random.randint(61,99) for i in range(0,5)]
    test2 = [random.randint(61,99) for i in range(0,5)]
    schoolset=list(zip(school_code,student_codes,test1,test2))
    df = pd.DataFrame(schoolset, columns=['school_code','student_code','test1','test2'])
    return df


# combine the mock dataframes into a master set
students_df=pd.DataFrame([])
for school_code in [500,600,700,800,900]:
    df = make_school(school_code)
    students_df = pd.concat([students_df, df])
# clean up the index   
students_df=students_df.reset_index(drop=True)
students_df
```

```python
# find the highest score for each student
students_df['hi_score']=students_df[['test1','test2']].max(axis=1)
students_df.sample(5)
```

```python
# what is the highest test score in each school?
schools_df_hi=students_df.groupby('school_code')[['school_code','hi_score']].max()
schools_df_hi
```

```python
# what is the student with the highest test score in each school?
topkidslist=[]
for school_code in [500,600,700,800,900]:
    school=students_df[students_df['school_code']==school_code]
    topkidslist.append(school.loc[school['hi_score'].idxmax()])
topkids=pd.DataFrame(topkidslist)
topkids
```
